Pasos:

- Busca el artista por nombre usando el endpoint /v1/search para obtener su artist_id. **hecho**

- Para cada álbum, llama al endpoint /v1/albums/{id}/tracks para obtener las canciones.
- Guarda o muestra los resultados, combinando el nombre del álbum y el de cada canción.

```python
# Lllamada al endpoint del La respuesta es un JSON
artista_json = sp.search(q='artist:Oasis')
artista_json
print(json.dumps(artista_json, indent=4))
```

```python
# Navegamos por el JSON usando clave-valor hasta llegar al id del artista
id_artista = artista_json["tracks"]["items"][0]["artists"][0]["id"]
id_artista
```

```python
albums = sp.artist_albums(id_artista, album_type="album", limit=50)
```

Ahora hay que sacar los albumes a partir del id del artista

```python
albums.keys()
```

```python
# Aquí interesa iterar, interesa sacar la info de todos
print(len(albums["items"]))
```

```python
# Aquí interesa iterar, interesa sacar la info de todos
lista_albumns = []

# En cada iteración se hace una llamada a la API
for i in albums["items"]:
    id_album = i["id"]
    name_album = i["name"]
    canciones = sp.album_tracks(id_album)
    # En cada iteracion devuelve la lista de tracks de cada album
    # print(canciones)ç
    for j in canciones["items"]:
        print(j["name"])
        lista_albumns.append({
            "nombre_artista": "Oasis",
            "nombre_disco": name_album,
            "nombre cancion": j["name"]
        })
```

```python
print(len(lista_albumns))
```

```python
df_canciones = pd.DataFrame(lista_albumns)
df_canciones
```

```python
df_canciones.to_csv("canciones.csv", sep = ';', index = True)
```

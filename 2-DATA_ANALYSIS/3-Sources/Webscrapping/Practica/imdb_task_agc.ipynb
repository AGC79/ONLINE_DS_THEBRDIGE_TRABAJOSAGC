## Web scrapping de IMDB

Descarga la información correspondiente y guarda en un csv el top de las 250 películas mediante webscrapping. Encapsúlalo en un script.

Obtén:
* Título
* Año
* Duración
* Posición
* Rating

# SCRIPT FINAL

### OJO: EN LA CORRECIÓN SE USA BY CSS SELECTOR, ES MEJOR QUE POR RUTA

```python
# La funcion find_element saca un solo elemento
# La funcion find_elements saca una lista de elementos
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import urllib3 # urllib3 es un cliente HTTP potente y fácil de usar para Python.
import re # Expresiones regulares 
import time
import pandas as pd

# Base para ejecutar el navegador
service = Service(executable_path='./chromedriver.exe')
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.imdb.com/chart/top/")

# boton_aceptar = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div/div[2]/div/button[2]')

boton_aceptar = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="__next"]/div[1]/div/div[2]/div/button[2]')))

boton_aceptar.click()

# listado_peliculas = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div/div[3]/section/div/div[2]/div/ul/li')

lista_top = []

listado_peliculas = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="__next"]/main/div/div[3]/section/div/div[2]/div/ul/li')))

for pelicula in listado_peliculas:
    
    posicion = pelicula.find_element(By.XPATH, ".//div[contains(@class, 'ipc-signpost__text')]").text
    posicion = int(posicion.replace("#", ""))
    # print(posicion)
    titulo = pelicula.find_element(By.XPATH, ".//h3[contains(@class, 'ipc-title__text')]").text
    # print(titulo)
    anio = int(pelicula.find_element(By.XPATH, ".//span[contains(@class, 'cli-title-metadata-item')][1]").text)
    # print(anio)
    duracion = pelicula.find_element(By.XPATH, ".//span[contains(@class, 'cli-title-metadata-item')][2]").text
    # print(duracion)
    rating = float(pelicula.find_element(By.XPATH, ".//span[contains(@class, 'ipc-rating-star--rating')]").text.replace(",", "."))
    # print(rating)

    lista_top.append({"Ranking": posicion,
                      "Título": titulo,
                      "Año": anio,
                      "Duración": duracion,
                      "Rating": rating
                      })

df_peliculas = pd.DataFrame(lista_top)

df_peliculas.to_csv("top_peliculas.csv", sep=";", index=True)

driver.quit()
```

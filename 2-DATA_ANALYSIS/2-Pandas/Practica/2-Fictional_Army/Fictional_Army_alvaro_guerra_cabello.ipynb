## CON ESTE TIPO DE MÁSCARAS NOS PODREMOS MOVER POR TODOS LO DATAFRAMES

A la mascara le meteremos las condiciones necesarias para encontra los datos.

```python
# Aquí estas mostrando todas las filas donde hay un null
df_army[(df_army["deaths"].isnull()) | (df_army["readiness"].isnull())]
```

```python
# Devuelve los registros en los que hay al menos un NaN en al menos una de sus columnas
df_army[df_army.isnull().any(axis=1)]
```

```python
# Hay 12 filas
# sacamos el 70% de la longitud de filas
len(df_army) * 0.7
```

```python
# modificar el DataFrame directamente (sin crear una copia nueva):
# esto borra la columna readiness, que presentana mas de un 30% de valores no validos
df_army.dropna(axis=1, thresh=len(df_army)*0.7, inplace=True)
```

```python
df_army = df_army.dropna(axis=1, thresh=len(df_army)*0.7)
df_army
```

### Step 3. Drop features, if there are more than 30% missing values.

```python
# Elimina las columnas donde al menos  el 30% de las filas tienen un NaN
# Es decir, mantiene aquellas en los que, al menos, un 70% de las filas es un valor válido, no NaN
df_army = df_army.dropna(axis=1, thresh=len(df_army)*0.7)
df_army
```

```python
df_army.isnull().sum()
```

### Step 4. Fill missing values with the mean of their regiment.

```python
df_army.groupby("regiment")["deaths"].mean()
```

```python
df_army.groupby("regiment")["deaths"].transform("mean")
```

```python
df_army["deaths"] = round(df_army["deaths"].fillna(df_army.groupby("regiment")["deaths"].transform("mean")), 1)
df_army
```

```python
# aquí el loc me sobra, ya que Scouts es un índice explicito
print(round(df_army.groupby("regiment")["deaths"].mean(), 1))
media_scouts = round(df_army.groupby("regiment")["deaths"].mean(), 1).loc["Scouts"]
media_scouts
```

```python
df_army = df_army.fillna(media_scouts)
df_army
```

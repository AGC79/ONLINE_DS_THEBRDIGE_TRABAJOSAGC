### Step 2. Create the 3 DataFrames based on the following raw data

```python
# Estos son 3 diccionarios
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

```

```python
import numpy as np
import pandas as pd
```

### Step 3. Assign each to a variable called data1, data2, data3

```python
# Dataframe 1
df_data1 = pd.DataFrame(raw_data_1)
df_data1
```

```python
# Dataframe 2
df_data2 = pd.DataFrame(raw_data_2)
df_data2
```

```python
# Dataframe 3
df_data3 = pd.DataFrame(raw_data_3)
df_data3
```

### Step 4. Join the two first dataframes along rows and assign to all_data

```python
# Join along rows
# axis es 0 por defecto
# vemos que el índice implicito se reinicia, con ignore_index=True tendremos el indice completo
df_all_data = pd.concat([df_data1, df_data2], ignore_index=True)
df_all_data
```

```python
df_all_data2 = pd.concat([df_data1, df_data2], join="outer")
df_all_data2
```

### Step 5. Join the two first dataframes along columns and assing to all_data_col

```python
# Join along columns
# Tenemos todas las columnas dup
df_all_data_col = pd.concat([
    df_data1.rename(columns={
        "first_name": "first_name_X",
        "subject_id": "subject_id_X",
        "last_name": "last_name_X"
    }),
    df_data2
], axis=1)

df_all_data
```

### Step 7. Merge all_data and data3 along the subject_id value

```python
# No haria falta, por defect coge la unica columna comun, on='subject_id'
# No haria falta tampoco how, ya que por defecto es inner
df_all_data_final = pd.merge(df_all_data, df_data3, on='subject_id', how="inner")
df_all_data_final
```

### Step 8. Merge only the data that has the same 'subject_id' on both data1 and data2

```python
# Vemos de inicio que tenemos diferentes alumnos con el mismo subject_id en dataframes diferentes
```

### Step 9. Merge all values in data1 and data2, with matching records and not matching records from both sides where available.

```python
# indicator=True, argumento que nos ayuda a identificar de dónde viene cada fila
df_merged_df1_df2_all = pd.merge(df_data1, df_data2, 
                                 on='subject_id', how="outer", 
                                 suffixes=["Vitoria'", "Bilbao"],
                                 indicator=True)
df_merged_df1_df2_all
```

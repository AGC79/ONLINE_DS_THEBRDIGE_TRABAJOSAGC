# Exploratory Data Analysis with Seaborn
Entender cómo se hacen las gráficas en python está muy bien, pero lo importante es saber cuándo usarlas. En este notebook no se pretende explicar todos los tipos de gráficas que hay (histogramas, scatter plots, heatmaps...), en primer lugar porque es imposible abarcarlo todo, y en segundo lugar porque hay que ir un paso atrás antes de elegir la gráfica, ya que tenemos que comprender de qué tipo son los datos y qué es lo que estamos buscando con cada gráfica.

# Tipos de las variables
Antes de representar cualquier variable es imprescindible comprender qué tipo de datos queremos representar. Tenemos los siguientes:

![imagen](./img/dtypes.png)

1. variabes cualitativas - son las variables clasiicativas o categóricas - catalogan (caro - medio - barato)

- nominales: son más exclusivas 
- ordinarias: no dan pistas sobre las variables, son los indices

1. variabes cuantitativas - son las variables de regresión, con las que se usa funciones de agregacion.
- discretas - valores que no suben o bajan tanto: número de estudiantes en una casa.
- continuas - velocidad son las variables de regresión: precio de coche, precio de casas

# Correlaciones en los Datos

Las correlaciones son **medidas estadísticas** que indican la **relación lineal** entre dos variables numéricas. Los valores de correlación varían entre **-1 y 1**:

- **1**: correlación positiva perfecta (cuando una variable aumenta, la otra también aumenta).
- **0**: no hay correlación lineal.
- **-1**: correlación negativa perfecta (cuando una variable aumenta, la otra disminuye).

## Interpretación de las correlaciones en los datasets

Por ejemplo, en el dataset de **Boston Housing**:

| Variable | Descripción |
|----------|------------|
| CRIM     | Tasa de criminalidad por ciudad |
| ZN       | Porcentaje de terrenos residenciales para lotes grandes |
| INDUS    | Proporción de acres comerciales no minoristas |
| CHAS     | Variable ficticia Charles River (1 si el tramo limita con el río, 0 de lo contrario) |
| NOX      | Concentración de óxidos de nitrógeno (ppm) |
| RM       | Número promedio de habitaciones por vivienda |
| AGE      | Porcentaje de unidades ocupadas por propietarios construidas antes de 1940 |
| DIS      | Distancia ponderada a cinco centros de empleo en Boston |
| RAD      | Índice de accesibilidad a carreteras radiales |
| TAX      | Tasa de impuesto a la propiedad |
| PTRATIO  | Ratio alumno/profesor por ciudad |
| B        | 1000(Bk - 0.63)^2, donde Bk es la proporción de afroamericanos por ciudad |
| LSTAT    | % de población de menor estatus socioeconómico |
| MEDV     | Valor medio de las viviendas ocupadas por propietarios (en $1000) |

En la matriz de correlación:

- Valores cercanos a **1** indican que las variables aumentan juntas. Por ejemplo, `RAD` y `TAX` tienen una correlación alta positiva (~0.91), lo que indica que a más accesibilidad a carreteras radiales, mayor es el impuesto a la propiedad.
- Valores cercanos a **-1** indican que las variables se mueven en direcciones opuestas. Por ejemplo, `LSTAT` y `MEDV` tienen una correlación negativa (~-0.74), indicando que a mayor % de población de menor estatus socioeconómico, menor es el valor de las viviendas.
- Valores cercanos a **0** indican poca o ninguna relación lineal entre variables. Por ejemplo, `CHAS` y `INDUS` tienen una correlación baja (~0.06).

## Uso de la correlación

- Identificar variables que influyen en otras (por ejemplo, factores que afectan el precio de viviendas).
- Detectar relaciones lineales antes de aplicar modelos estadísticos o de Machine Learning.
- Explorar tendencias y patrones en los datos de manera rápida.

> **Nota:** La correlación **no implica causalidad**. Dos variables pueden estar correlacionadas sin que una cause a la otra.


# Correlación en el dataset Boston

El comando `boston_df.corr()` de **pandas** calcula la **matriz de correlación** de todas las columnas numéricas del DataFrame `boston_df`.

Por defecto, utiliza el **coeficiente de correlación de Pearson**, que mide la **relación lineal** entre dos variables.

## Interpretación rápida

- Valores cercanos a `1`: correlación positiva fuerte (ambas variables suben juntas).  
- Valores cercanos a `-1`: correlación negativa fuerte (cuando una sube, la otra baja).  
- Valores cercanos a `0`: casi ninguna correlación lineal.  

## Ejemplo de salida simplificada

|        | CRIM  | RM    | MEDV  |
|--------|-------|-------|-------|
| CRIM   | 1.00  | -0.22 | -0.39 |
| RM     | -0.22 | 1.00  | 0.70  |
| MEDV   | -0.39 | 0.70  | 1.00  |

- `RM` y `MEDV` tienen correlación positiva alta (~0.70) → más habitaciones, más valor de vivienda.  
- `CRIM` y `MEDV` tienen correlación negativa (~-0.39) → más criminalidad, menor valor de vivienda.

> Si quieres, puedes visualizar esta matriz con un **heatmap** para interpretarla mejor.


```python
boston_df.corr()
```

La correlaciones pueden ser positivas o negativas. Este tipo de gráficos son muy buenos para modelos predictivos.

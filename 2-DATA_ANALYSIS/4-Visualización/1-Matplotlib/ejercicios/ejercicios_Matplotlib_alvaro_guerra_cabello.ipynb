## Ecuaciones de la recta

Una recta en el plano puede escribirse de dos formas principales:

---

### 1) Forma Punto–Pendiente

\[
y - y_1 = m(x - x_1)
\]

Se utiliza cuando conocemos:
- La **pendiente**: \(m\)
- Un **punto por donde pasa la recta**: \((x_1, y_1)\)

| Símbolo | Significado |
|--------|-------------|
| \(m\) | Pendiente (inclinación de la recta) |
| \((x_1, y_1)\) | Punto conocido sobre la recta |

---

### 2) Forma Pendiente–Intersección

\[
y = mx + b
\]

Se utiliza cuando conocemos:
- La **pendiente**: \(m\)
- El **corte con el eje Y** (ordenada de origen): \(b\)

| Símbolo | Significado |
|--------|-------------|
| \(m\) | Pendiente |
| \(b\) | Valor de \(y\) cuando \(x = 0\) (punto de corte con el eje Y) |

El punto \((0, b)\) es donde la recta toca el **eje Y**.

---

### Relación entre ambas formas

La forma punto–pendiente puede convertirse en la forma pendiente–intersección simplemente despejando \(y\):

$$
y - y_1 = m(x - x_1) \;\Rightarrow\; y = mx + b
$$

Es decir, **son la misma recta**, solo escrita de forma diferente.

---

### Interpretación visual

- Si \(m > 0\): la recta **sube** hacia la derecha.
- Si \(m < 0\): la recta **baja** hacia la derecha.
- \(b\) indica dónde la recta toca el eje Y.





```python
x = np.linspace(0, 5, 100) # 100 es la precisión en el eje x

plt.figure()

plt.subplot(1, 2, 1)
plt.plot(x, x * 3, linestyle='-', color='blue');

plt.subplot(1, 2, 2) 
plt.plot(x, (x * (-3)) - 5, linestyle='--', color='red');
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

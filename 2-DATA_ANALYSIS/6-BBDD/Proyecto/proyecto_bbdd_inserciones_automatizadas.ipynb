```python
# Convertir fecha a formato YYYY-MM-DD
df_alumnos_ds_total['Fecha_comienzo'] = pd.to_datetime(
    df_alumnos_ds_total['Fecha_comienzo'], 
    dayfirst=True
).dt.date

# Mapas de IDs
promocion_map = {
    'Febrero': 0,
    'Marzo': 1,
    'Abril': 2,
    'Junio': 3,
    'Septiembre': 4,
    'Octubre': 5
}

campus_map = {
    'Madrid': 0,
    'Valencia': 1,
    'Bilbao': 2,
    'Vitoria': 3,
    'Sevilla': 4
}

bootcamp_map = {
    'DS': 0,
    'FS': 1
}

# Como todos los proyectos actuales son de DS, asignamos DS
df_alumnos_ds_total['id_bootcamp'] = 0  # DS

# Crear DataFrame final de alumnos único
df_final_alumnos_ds = df_alumnos_ds_total[['Nombre','Email','Promoción','Campus','Fecha_comienzo','id_bootcamp']].drop_duplicates(subset=['Nombre','Email'])

# Asignar IDs
df_final_alumnos_ds['id_alumno'] = range(len(df_final_alumnos_ds))
df_final_alumnos_ds['id_promocion'] = df_final_alumnos_ds['Promoción'].map(promocion_map)
df_final_alumnos_ds['id_campus'] = df_final_alumnos_ds['Campus'].map(campus_map)

# Renombrar columnas
df_final_alumnos_ds = df_final_alumnos_ds.rename(columns={
    'Nombre': 'nombre',
    'Email': 'email',
    'Fecha_comienzo': 'fecha_inicio'
})

# Seleccionar columnas finales
df_final_alumnos_ds = df_final_alumnos_ds[['id_alumno','nombre','email','id_bootcamp','id_promocion','id_campus','fecha_inicio']]

df_final_alumnos_ds
```

```python
df_final_alumnos_ds.to_sql("alumno", engine, if_exists="append", index= False)
```

```python
df_alumnos_fs_01 = pd.read_csv(".\\data\\clase_3.csv", sep=";")
df_alumnos_fs_01 
```

```python
df_alumnos_fs_02 = pd.read_csv(".\\data\\clase_4.csv", sep=";")
df_alumnos_fs_02 
```

```python
df_alumnos_fs_total = df_alumnos = pd.concat([df_alumnos_fs_01, df_alumnos_fs_02], ignore_index=True)
df_alumnos_fs_total
```

```python
# Convertir fecha a formato YYYY-MM-DD
df_alumnos_fs_total['Fecha_comienzo'] = pd.to_datetime(
    df_alumnos_fs_total['Fecha_comienzo'], 
    dayfirst=True
).dt.date

# Seleccionar columnas únicas por alumno (sin proyectos)
df_final_alumnos_fs = df_alumnos_fs_total[['Nombre','Email','Promoción','Campus','Fecha_comienzo']].drop_duplicates()

# Crear id_alumno único empezando después de DS
start_id = df_final_alumnos_ds['id_alumno'].max() + 1
df_final_alumnos_fs['id_alumno'] = range(start_id, start_id + len(df_final_alumnos_fs))

# Mapear IDs de bootcamp, promoción y campus
df_final_alumnos_fs['id_bootcamp'] = 1  # FS
df_final_alumnos_fs['id_promocion'] = df_final_alumnos_fs['Promoción'].map(promocion_map)
df_final_alumnos_fs['id_campus'] = df_final_alumnos_fs['Campus'].map(campus_map)

# Renombrar columnas
df_final_alumnos_fs = df_final_alumnos_fs.rename(columns={
    'Nombre': 'nombre',
    'Email': 'email',
    'Fecha_comienzo': 'fecha_inicio'
})

# Seleccionar columnas finales
df_final_alumnos_fs = df_final_alumnos_fs[['id_alumno','nombre','email','id_bootcamp','id_promocion','id_campus','fecha_inicio']]

df_final_alumnos_fs

```

```python
# Insertar en la tabla
df_final_alumnos_fs.to_sql("alumno", engine, if_exists="append", index=False)
```

```python
# --- DS ---
# Columnas de proyectos DS
proyectos_ds_cols = ['Proyecto_HLF', 'Proyecto_EDA', 'Proyecto_BBDD', 'Proyecto_ML', 'Proyecto_Deployment']
# IDs correspondientes
proyectos_ds_ids = [0, 1, 2, 3, 4]

records_ds = []
id_proyecto_realizado = 0

for _, row in df_alumnos_ds_total.iterrows():
    id_alumno = df_final_alumnos_ds.loc[
        (df_final_alumnos_ds['nombre'] == row['Nombre']) & 
        (df_final_alumnos_ds['email'] == row['Email']), 'id_alumno'
    ].values[0]
    
    for col, id_proyecto in zip(proyectos_ds_cols, proyectos_ds_ids):
        calificacion = row[col]
        records_ds.append({
            'id_proyectos_realizados': id_proyecto_realizado,
            'id_alumno': id_alumno,
            'id_proyecto': id_proyecto,
            'calificacion': calificacion
        })
        id_proyecto_realizado += 1

df_proyectos_ds = pd.DataFrame(records_ds)


# --- FS ---
# Columnas de proyectos FS
proyectos_fs_cols = ['Proyecto_WebDev', 'Proyecto_FrontEnd', 'Proyecto_Backend', 'Proyecto_React', 'Proyecto_FullSatck']
# IDs correspondientes
proyectos_fs_ids = [5, 6, 7, 8, 9]

records_fs = []

for _, row in df_alumnos_fs_total.iterrows():
    id_alumno = df_final_alumnos_fs.loc[
        (df_final_alumnos_fs['nombre'] == row['Nombre']) & 
        (df_final_alumnos_fs['email'] == row['Email']), 'id_alumno'
    ].values[0]
    
    for col, id_proyecto in zip(proyectos_fs_cols, proyectos_fs_ids):
        calificacion = row[col]
        records_fs.append({
            'id_proyectos_realizados': id_proyecto_realizado,
            'id_alumno': id_alumno,
            'id_proyecto': id_proyecto,
            'calificacion': calificacion
        })
        id_proyecto_realizado += 1

df_proyectos_fs = pd.DataFrame(records_fs)

# Concatenar DS y FS
df_proyectos_realizados = pd.concat([df_proyectos_ds, df_proyectos_fs], ignore_index=True)
```

```python
df_proyectos_realizados.to_sql("proyectos_realizados", engine, if_exists="append", index=False)
```
